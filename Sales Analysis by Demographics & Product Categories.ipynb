# Exploratory Data Analysis

# Gender

```python
plt.figure(figsize=(4,4))  
ax = sns.countplot(x='Gender', data=df, hue='Gender')  

ax.bar_label(ax.containers[0]) 
ax.bar_label(ax.containers[1])  

plt.title("Gender Distribution of Buyers")
plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(6,6))
ax = sns.barplot(x='Gender', y='Amount', data=sales_gen, hue = 'Gender')

ax.bar_label(ax.containers[0], fmt='%.0f', label_type='edge', padding=3)

plt.title("Total Sales Amount by Gender")
plt.show()
```

#From the graph above, we can see that most buyers are female compared to men, and the purchasing power of females is also greater than that of men.From the graph above, we can see that most buyers are female compared to men, and the purchasing power of females is also greater than that of men.

# Age

```python
plt.figure(figsize = (7,7))
ax = sns.countplot(x = "Age Group" , data = df , hue = "Gender")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Most buyers from both genders belong to different age groups.")
plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen , hue = "Gender")

plt.title("Total Sales Amount by Age Group")
plt.show()
```

#From above graphs we can see that most of the buyers are of age group between 26-35 yrs female.

# State

```python
#1. Total Number of orders from top 10 states.
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
plt.figure(figsize = (7,7))
ax = sns.barplot(x = 'State' , y = 'Orders' , data = sales_state)
ax.bar_label(ax.containers[0])
plt.xticks(rotation = 90)
plt.title("Total Orders from Top 10 States")
plt.show()
```

```python
#2. Total Amount/Sales from Top 10 States.
sales_state = df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize = (10,10))
ax = sns.barplot(x = 'State' , y = 'Amount' , data = sales_state)
ax.bar_label(ax.containers[0], fmt='%.0f', label_type='edge', padding=3)
plt.xticks(rotation = 90)
plt.title("Total Amount or Sales from Top 10 States")
plt.show()
```

#From the graph above, we can see that, unexpectedly, most of the orders are from Uttar Pradesh, Maharashtra, and Karnataka, respectively. However, the total sales amount is highest in Uttar Pradesh, followed by Karnataka and Maharashtra.

# Marital Status

```python
plt.figure(figsize=(4,4))  
ax = sns.countplot(x='Marital Status', data=df)  

ax.bar_label(ax.containers[0]) 

plt.title("Count of Buyers by Marital Status")
plt.show()
```

```python
sales_state = df.groupby(['Marital Status' , 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ax = sns.countplot(x = "Marital Status" , data = df , hue = "Gender")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Purchase Distribution by Marital Status and Gender")
plt.show()
```

#From the graph above, we can see that most buyers are married women, and they have high purchasing power.

# Occupation

```python
plt.figure(figsize=(10,10))  
ax = sns.countplot(x='Occupation', data=df)  

ax.bar_label(ax.containers[0]) 

plt.xticks(rotation = 90)
plt.title("Count of Buyers by Occupation")
plt.show()
```

```python
sales_state = df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize = (14,14))
ax = sns.barplot(x = 'Occupation' , y = 'Amount' , data = sales_state)
ax.bar_label(ax.containers[0], fmt='%.0f', label_type='edge', padding=3)
plt.xticks(rotation = 90)
plt.title("Total Sales Amount by Occupation")
plt.show()
```

#From Above Graph we can see that most of the buyers are working in IT, Aviation and Healthcare Sector.

# Product Category

```python
plt.figure(figsize=(10,10))  
ax = sns.countplot(x='Product_Category', data=df)  

ax.bar_label(ax.containers[0]) 

plt.xticks(rotation = 90)
plt.title("Count of Sold Products by Category")
plt.show()
```

```python
sales_product = df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize = (17,17))
ax = sns.barplot(x = 'Product_Category' , y = 'Amount' , data = sales_product)
ax.bar_label(ax.containers[0], fmt='%.0f', label_type='edge', padding=3)
plt.xticks(rotation = 90)
plt.title("Total Sales Amount by Top 10 Product Categories")
plt.show()
```

#From the graph above, we can see that most of the sold products belong to the Food, Footwear, and Electronics categories.

```python
sales_product = df.groupby('Product_ID', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize = (17,17))
ax = sns.barplot(x = 'Product_ID' , y = 'Amount' , data = sales_product)
ax.bar_label(ax.containers[0], fmt='%.0f', label_type='edge', padding=3)
plt.xticks(rotation = 90)
plt.title("Total Sales Amount by Top 10 Products")
plt.show()
```

# Conclusion

#Married individuals aged 26-35 from Uttar Pradesh, Maharashtra, and Karnataka, working in the IT, Healthcare, and Aviation sectors, are more likely to purchase products from the Food, Clothing, and Electronics categories.
